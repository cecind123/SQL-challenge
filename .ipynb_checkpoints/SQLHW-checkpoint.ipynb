```python
import pandas as pd
import matplotlib.pyplot as plt
#%matplotlib inline

from sqlalchemy import create_engine
engine = create_engine('postgresql://postgres:postgres@localhost:5432/SQLHW')
connection = engine.connect()
```

```python
#Read in Employee data to DF
employees = pd.read_sql('select * from Employees', connection)
employees.head()
```

```python
#Read in Salary data into DF
salaries = pd.read_sql('select emp_no, salary from Salaries', connection)
salaries.head()
```

```python
#Merge DFs
merged_df = pd.merge(employees,salaries,on="emp_no")
merged_df.head()
```

```python
#Read in job title data into DF
titles = pd.read_sql('select * from Titles',connection)
titles.head()
```

```python
#Merge DFs again to get DF with salary and job titles
merged_df_2 = pd.merge(merged_df,titles, on="emp_no")
merged_df_2.head()
```

```python
#Group by job title and find avg salaries and make grouped DF
grouped_df = pd.DataFrame(merged_df_2.groupby("title").mean())
grouped_df = grouped_df.drop(columns="emp_no")

grouped_df
```

```python
#Plot bar chart of Avg Salary by Job Title
grouped_df.plot(kind='bar')
plt.title(label="Average Salary by Title")
plt.ylabel("Average Salary")
plt.xlabel("Job Titles")
plt.savefig("Average_Salary_By_Title.png", bbox_inches="tight")
plt.show()
```
